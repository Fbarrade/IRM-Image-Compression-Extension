```python
import PIL
from matplotlib import pyplot as plt
import numpy as np
import IRM as irm
```

```python
im=PIL.Image.open(r"images\tokyo.bmp")
img=np.array(im)
```

```python
#compressing the image 
irm.irmCode(r"images\tokyo.bmp",0)  # 0 : lossless
```

```python
#decompresser le fichier irm
imdeco = irm.irmDECO(r"images\tokyo_lossless.irm")
```

```python
#affichage de l'image compressée et l original : 
irm.showtwo(imdeco,img) 
```

```python
print("----------------- image  : tokyo 500x300------------------") 
print()

#taux:
print("le taux lossless: ",irm.taux("images/tokyo_lossless.irm","images/tokyo.bmp"),"%")
irm.irmCode("images/tokyo.bmp",1)
print("le taux lossy: quant ",irm.taux("images/tokyo_lossy.irm","images/tokyo.bmp"),"%")
irm.irmCode("images/tokyo.bmp",2)
print("le taux lossy: hls ",irm.taux("images/tokyo_lossy.irm","images/tokyo.bmp"),"%")
irm.irmCode("images/tokyo.bmp",3)
print("le taux lossy: hls+quant ",irm.taux("images/tokyo_lossy.irm","images/tokyo.bmp"),"%")

print()

#MSE
print("MSE lossless: ",irm.MSE("images/tokyo_lossless.irm","images/tokyo.bmp"))
irm.irmCode("images/tokyo.bmp",1)
print("MSE lossy: quant ",irm.MSE("images/tokyo_lossy.irm","images/tokyo.bmp"))
irm.irmCode("images/tokyo.bmp",2)
print("MSE lossy: hls",irm.MSE("images/tokyo_lossy.irm","images/tokyo.bmp"))
irm.irmCode("images/tokyo.bmp",3)
print("MSE lossy: hls+quant",irm.MSE("images/tokyo_lossy.irm","images/tokyo.bmp"))

#entropie 
print()
print("l'entropie :",irm.entropie("images/tokyo.bmp"))

#efficacite 
print()
print("L'efficacité est:",(irm.efficacite("images/tokyo.bmp")),"%")
print()

#Rendement 
print("le rendement ",irm.rendement("images/tokyo.bmp")*100,"%")
print()

#Redondance
print("la redondance :",irm.redondance("images/tokyo.bmp"))

plt.imshow(im)
```
